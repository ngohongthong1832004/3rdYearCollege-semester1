```python
df_3tinh = df[df['area'].apply(lambda x : x in ['TP HCM','Hà Nội','Đà Nẵng'])]
print("Số dòng của dữ liệu mới là  : ",df_3tinh.shape[0])
df_3tinh['area'].value_counts()
```

```python
(df_3tinh.describe().T)
```

```python
sosanh = (df_3tinh.describe().T)[['mean','50%','std']]
sosanh = sosanh.rename(columns={"50%": "median"})
sosanh
```

```python
sosanh.plot(kind='bar',figsize=(10,5),title='So sanh gia tri trung binh, trung vi, do lech chuan cua 3 tinh',rot=0)
plt.show()
```

```python
plt.bar(x= ['mean','median','std'], height = [df_3tinh['km'].mean(),df_3tinh['km'].median(),df_3tinh['km'].std()])
plt.show()
```

```python
# Vẽ theo màu xe
df_color = df.groupby('in_color').count() # Group by ngoại thất color

df_color.sort_values(by='car_model',inplace=True)

ax = df_color[-10:].plot.pie(y='car_model',legend=3,figsize=(8, 8), title='Tỷ lệ màu sắc',autopct='%.2f')
ax.set_ylabel('')
```

Ta có thể thấy số xe màu đen , màu Kem chiếm hơn 50% loại màu

```python
xecu = df[df['new_old']=='Xe cũ']
xecu.head()
```

```python
plt.scatter(xecu['km'],xecu['price'])
plt.ylabel("Gía xe")
plt.xlabel('Số km')
plt.show()
```

Hai đại lượng này có vẻ như không có quan hệ vì ta thấy nhiều xe cùng giá xe nhưng số km có thể khác nhau nhiều

```python
quanhe = df[df['car_type'].apply(lambda x : x in ['SUV','Sedan'])]
```

```python

giaxe = quanhe[['car_type','price','car_year','car_model']].groupby(['car_type','car_year']).sum()
giaxe
```

```python
giaxe = giaxe.reset_index()

giaxe = giaxe.pivot(index='car_year', columns='car_type', values='price')
```

```python
giaxe.plot(kind='bar', figsize=(10, 6))
plt.show()
```

Nhận xét : 
- Giá xe của SUV qua các năm thường cao hơn Sedan
- Nhưng năm từ 1990 - 2006 thì gần như giá xe bằng nhau
- Từ 2019 đến nay thì SUV cao hơn hẳn
