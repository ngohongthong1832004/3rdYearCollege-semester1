#### 1. Mức lương thay đổi qua các năm như thế nào giữa những nhóm người khác nhau?



````python
print("Câu trả lời cần làm là : ",(56 % 3) + 1)
````

````python
# YOUR CODE HERE
# Tạo cột phân loại nhóm (Chief/Non-Chief)
df['Group'] = df['JobTitle'].apply(lambda x: 'Chief' if 'Chief' in x else 'Non-Chief')

# Tính mức lương trung bình theo nhóm và năm
salary_by_group_year = df.groupby(['Year', 'Group'])['TotalPay'].mean().unstack()

# Vẽ biểu đồ đường
plt.figure(figsize=(10, 6))

for group in salary_by_group_year.columns:
    plt.plot(salary_by_group_year.index, salary_by_group_year[group], marker='o', label=group)

# Thêm tiêu đề và các nhãn trục
plt.title('Mức lương trung bình qua các năm giữa các nhóm (Chief và Non-Chief)', fontsize=16)
plt.xlabel('Năm', fontsize=14)
plt.ylabel('Mức lương trung bình (USD)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)

# Thêm chú giải (legend)
plt.legend(loc='upper left', fontsize=10, title='Nhóm')

plt.tight_layout()
plt.show()
````

#### 3. Cho biết có bao nhiêu người là sếp (có từ `Chief`) trong Job Title



````python
# YOUR CODE HERE
# Lọc các JobTitle chứa từ 'Chief'
chief_count = df[df['JobTitle'].str.contains('Chief', case=False, na=False)].shape[0]

# In kết quả
print("Số lượng người có từ 'Chief' trong Job Title: ", chief_count)
````

````python
# Lọc các JobTitle chứa từ 'Chief'
chief_count = df[df['JobTitle'].str.contains('Chief', case=False, na=False)].shape[0]
non_chief_count = df.shape[0] - chief_count

# Dữ liệu cho biểu đồ
labels = ['Chief', 'Non-Chief']
sizes = [chief_count, non_chief_count]
colors = ['#ff9999','#66b3ff']
explode = (0.1, 0)  # tách phần 'Chief' ra khỏi biểu đồ

# Vẽ biểu đồ tròn
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Tỷ lệ người có từ "Chief" trong Job Title')
plt.axis('equal')  # đảm bảo biểu đồ tròn không bị méo
plt.show()
````
