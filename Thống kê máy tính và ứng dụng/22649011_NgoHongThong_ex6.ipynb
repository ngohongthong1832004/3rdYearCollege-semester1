```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
```

## Bài 1

```python
POP = np.random.uniform(120, 200, 10000)

# a. 
mean_height = np.mean(POP)  # Chiều cao trung bình
std_height = np.std(POP)    # Độ lệch chuẩn của chiều cao

# b. 
tall_people_ratio = np.sum(POP >= 180) / len(POP)  # Tỷ lệ người có chiều cao >= 180 cm

# c. 
plt.figure(figsize=(10, 5))
plt.hist(POP, bins=30, alpha=0.7, edgecolor='black')
plt.title('Biểu đồ tần số phân bố chiều cao trong dân số')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Tần số')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# d. 
plt.figure(figsize=(10, 5))
plt.hist(POP, bins=30, density=True, alpha=0.7, edgecolor='black', label='Dữ liệu dân số')
x = np.linspace(120, 200, 1000)
pdf = stats.norm.pdf(x, loc=mean_height, scale=std_height)  # Hàm mật độ xác suất (PDF) cho phân phối chuẩn
plt.plot(x, pdf, 'r--', label='Phân phối chuẩn (tính từ dân số)')
plt.title('Biểu đồ tần số và đường cong phân phối chuẩn')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Mật độ')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

mean_height, std_height, tall_people_ratio
```

## Bài 2

```python
sample = np.random.choice(POP, 20)

# a. 
sample_mean = np.mean(sample)  # Trung bình mẫu
sample_std = np.std(sample)    # Độ lệch chuẩn mẫu

# b. 
sample_tall_ratio = np.sum(sample >= 180) / len(sample)  # Tỷ lệ người có chiều cao >= 180 cm trong mẫu

# c. 
plt.figure(figsize=(10, 5))
plt.hist(sample, bins=10, alpha=0.7, color='green', edgecolor='black')
plt.title('Biểu đồ tần số phân bố chiều cao của mẫu')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Tần số')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# d.
plt.figure(figsize=(10, 5))
plt.hist(sample, bins=10, density=True, alpha=0.7, color='green', edgecolor='black', label='Dữ liệu mẫu')
x = np.linspace(120, 200, 1000)
pdf_sample = stats.norm.pdf(x, loc=sample_mean, scale=sample_std)  # Hàm mật độ xác suất (PDF) cho phân phối chuẩn mẫu
plt.plot(x, pdf_sample, 'g--', label='Phân phối chuẩn (tính từ mẫu)')
plt.title('Biểu đồ tần số và đường cong phân phối chuẩn của mẫu')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Mật độ')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

sample_mean, sample_std, sample_tall_ratio
```

### Đồ thị của hàm mật độ phân phối chuẩn khác biệt đáng kể so với histogram của mẫu. Điều này là do kích thước mẫu nhỏ, dẫn đến độ lệch lớn và không đủ dữ liệu để tạo ra một phân phối chuẩn rõ ràng. Với kích thước mẫu lớn hơn, histogram của mẫu sẽ gần hơn với phân phối chuẩn đã vẽ.

## Bài 7

```python
weights = [2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8]

sample_mean = np.mean(weights)  # Trung bình mẫu
n = len(weights)  # Kích thước mẫu
confidence_level = 0.95  # Độ tin cậy 95%
alpha = 1 - confidence_level

# a. 
sigma = 0.3
z_score = stats.norm.ppf(1 - alpha / 2)  
margin_of_error_a = z_score * sigma / np.sqrt(n)  
confidence_interval_a = (sample_mean - margin_of_error_a, sample_mean + margin_of_error_a)

# b. 
sample_std = np.std(weights, ddof=1)  
t_score = stats.t.ppf(1 - alpha / 2, df=n-1) 
margin_of_error_b = t_score * sample_std / np.sqrt(n)  
confidence_interval_b = (sample_mean - margin_of_error_b, sample_mean + margin_of_error_b)

confidence_interval_a, confidence_interval_b
```

## Bài 8

```python
data = pd.read_csv("SAT-lard.csv")
data.head()
```

```python
# Lấy cột SAT score
sat_scores = data['SAT Score']

# a. 
sample_mean_a = sat_scores.mean()
sample_std_a = sat_scores.std()
n_a = len(sat_scores)
confidence_level_a = 0.99
alpha_a = 1 - confidence_level_a

z_score_a = stats.norm.ppf(1 - alpha_a / 2)
margin_of_error_a = z_score_a * sample_std_a / np.sqrt(n_a)
confidence_interval_a = (sample_mean_a - margin_of_error_a, sample_mean_a + margin_of_error_a)

# b. 
confidence_level_b = 0.95
alpha_b = 1 - confidence_level_b

z_score_b = stats.norm.ppf(1 - alpha_b / 2)
margin_of_error_b = z_score_b * sample_std_a / np.sqrt(n_a)
confidence_interval_b = (sample_mean_a - margin_of_error_b, sample_mean_a + margin_of_error_b)

# c. 
population_mean_c = sample_mean_a

# d. 
sample_36 = sat_scores.head(36)
sample_36_mean = sample_36.mean()
sample_36_std = sample_36.std()
n_36 = len(sample_36)
confidence_level_d = 0.99
alpha_d = 1 - confidence_level_d

t_score_d = stats.t.ppf(1 - alpha_d / 2, df=n_36 - 1)
margin_of_error_d = t_score_d * sample_36_std / np.sqrt(n_36)
confidence_interval_d = (sample_36_mean - margin_of_error_d, sample_36_mean + margin_of_error_d)

# e. 
confidence_level_e = 0.95
alpha_e = 1 - confidence_level_e

t_score_e = stats.t.ppf(1 - alpha_e / 2, df=n_36 - 1)
margin_of_error_e = t_score_e * sample_36_std / np.sqrt(n_36)   
confidence_interval_e = (sample_36_mean - margin_of_error_e, sample_36_mean + margin_of_error_e)

results = {
    "a. Khoảng tin cậy 99% (Toàn bộ dữ liệu)": confidence_interval_a,
    "b. Khoảng tin cậy 95% (Toàn bộ dữ liệu)": confidence_interval_b,
    "c. Trung bình điểm SAT của toàn trường": population_mean_c,
    "d. Khoảng tin cậy 99% (Mẫu 36 học sinh)": confidence_interval_d,
    "e. Khoảng tin cậy 95% (Mẫu 36 học sinh)": confidence_interval_e,
}

results
```
