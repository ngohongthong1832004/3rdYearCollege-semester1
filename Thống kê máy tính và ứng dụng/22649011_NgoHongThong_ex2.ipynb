## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
import numpy as np
arr = np.random.randint(0, 1000, 100)
print(arr)
print("-----------------")
print(arr[:10])
print("-----------------")
print(arr[-5:])
print("-----------------")
print(arr[(arr % 5 == 0) & (arr > 100)])
````

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

````python
print(len(arr[arr < 100]))
print("-----------------")
print(arr[arr < 100])
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
print(np.sum(arr))
print("-----------------")
print(np.mean(arr))
print("-----------------")
print(np.median(arr))
print("-----------------")
print(np.percentile(arr, 25))
print("-----------------")
print(np.percentile(arr, 50))
print("-----------------")
print(np.percentile(arr, 75))
print("-----------------")
print(np.percentile(arr, 20))
````

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

````python
for i in range(5):
    print(i + 1, " : " ,np.random.choice(arr, 10))
````

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

````python
print(np.bincount(arr).argmax())
````

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

````python
x = np.linspace(0, 2 * np.pi, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)', linestyle='--')

plt.title('Đồ thị')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
````

### Câu 5: Vẽ các đồ thị con

````python
x = np.linspace(0, 2 * np.pi, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 2, 1)
plt.plot(x, y_sin, label='sin(x)')
plt.title('Đồ thị')
plt.legend(loc="lower right")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(x, y_cos, label='cos(x)', color="red")
plt.title('Đồ thị')
plt.legend(loc="lower right")
plt.grid(True)

plt.tight_layout()
plt.show()

````

## SEABORN



Seaborn là một thư viện hỗ trợ trực quan hóa được viết dựa trên thư viện `matplotlib`. Trong bài tập này, bạn hãy tự tìm hiểu thư viện `seaborn` và trực quan hóa dữ liệu tiền tip tại một nhà hàng.

Bạn có thể dùng đoạn mã sau để đọc dữ liệu có sẵn trong `seaborn`

```python
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
```
Hoặc bạn có thể đọc dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)
- `tip`: tiền tip/ tiền boa (USD)
- `sex`: giới tính người thanh toán hóa đơn
- `smoker`: có hút thuốc hay không
- `day`: ngày trong tuần
- `time`: thời gian dùng bữa(buổi sáng, buổi trưa, buổi tối)
- `size`: số lượng người tham gia

Bạn hãy lựa chọn loại đồ thị phù hợp để trả lời những câu hỏi sau:

````python
import seaborn as sns

df = sns.load_dataset("tips")
df.head()
````

### 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày

````python
tip_by_day = df.groupby('day')['tip'].sum().reset_index()

sns.barplot(x='day', y='tip', data=tip_by_day)

plt.xlabel('Ngày')
plt.ylabel('Tiền bo')
plt.title('Tiền bo theo ngày')

plt.show()
````

### 2. Vẽ đồ thị thể hiện tổng tiền tip theo thời gian dùng bữa (time)


````python
tip_sum_by_time = df.groupby('time')['tip'].sum().reset_index()

sns.barplot(x='time', y='tip', data=tip_sum_by_time)

plt.xlabel('Bữa ăn')
plt.ylabel('Tiền bo')
plt.title('Tiền bo theo bữa ăn')

plt.show()
````

### 3. Vẽ đồ thị histogram của tổng hoá đơn


````python
sns.histplot(df['total_bill'], bins=5, kde=False)

# Thêm tiêu đề và nhãn
plt.title('Histogram của Tổng Hoá Đơn')
plt.xlabel('Tổng Hoá Đơn')
plt.ylabel('Số Lượng')

# Hiển thị đồ thị
plt.show()
````

### 4. Vẽ đồ thị boxplot của tổng hoá đơn


````python
sns.boxplot(x=df['total_bill'])

plt.title('Boxplot của Tổng Hoá Đơn')
plt.xlabel('Tổng Hoá Đơn')

plt.show()
````

### 5. Vẽ đồ thị boxplot so sánh tổng hóa đơn giữa các ngày trong tuần


````python
sns.boxplot(x='day', y='total_bill', data=df)

plt.title('Boxplot của Tổng Hoá Đơn Theo Ngày Trong Tuần')
plt.xlabel('Ngày')
plt.ylabel('Tổng Hoá Đơn')

plt.show()
````

### 6. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip. (`scatter plot`)


````python
sns.scatterplot(x='total_bill', y='tip', data=df)

plt.title('Sự Liên Quan Giữa Tổng Hoá Đơn và Tiền Tip')
plt.xlabel('Tổng Hoá Đơn')
plt.ylabel('Tiền Tip')

plt.show()
````

### 7. Tìm hiểu và sử dụng hàm `jointplot` trong seaborn để khảo sát quan hệ giữa hai thuộc tính `total_bill` và `tip`. Theo bạn hai thuộc tính này có tương quan với nhau hay không?


````python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter', color='blue')

# Thêm tiêu đề
plt.suptitle('Mối Quan Hệ Giữa Tổng Hoá Đơn và Tiền Tip', y=1.02)

# Hiển thị đồ thị
plt.show()
````

### 8. Tìm hiểu và sử dụng hàm `pairplot` trong seaborn để trực quan bộ dữ liệu. Theo bạn, trong bộ dữ liệu có những thuộc tính nào liên quan với nhau không?



````python
sns.pairplot(df, hue='sex', diag_kind='hist')

# Thêm tiêu đề
plt.suptitle('Trực Quan Bộ Dữ Liệu', y=1.02)

# Hiển thị đồ thị
plt.show()
````

### 9. Vẽ hai đồ thị histogram cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`



````python
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Vẽ histogram cho thời gian dùng bữa 'Dinner'
sns.histplot(df[df['time'] == 'Dinner']['total_bill'], bins=5, ax=axes[0], color='blue')
axes[0].set_title('Tổng Hoá Đơn - Dinner')
axes[0].set_xlabel('Tổng Hoá Đơn')
axes[0].set_ylabel('Số Lượng')

# Vẽ histogram cho thời gian dùng bữa 'Lunch'
sns.histplot(df[df['time'] == 'Lunch']['total_bill'], bins=5, ax=axes[1], color='green')
axes[1].set_title('Tổng Hoá Đơn - Lunch')
axes[1].set_xlabel('Tổng Hoá Đơn')

# Thêm tiêu đề cho toàn bộ figure
plt.suptitle('So Sánh Tổng Hoá Đơn Giữa Lunch và Dinner')

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
````

### 10. Vẽ hai đồ thị boxplot cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Vẽ boxplot cho thời gian dùng bữa 'Dinner'
sns.boxplot(x='time', y='total_bill', data=df[df['time'] == 'Dinner'], ax=axes[0], color='blue')
axes[0].set_title('Tổng Hoá Đơn - Dinner')
axes[0].set_xlabel('Thời Gian')
axes[0].set_ylabel('Tổng Hoá Đơn')

# Vẽ boxplot cho thời gian dùng bữa 'Lunch'
sns.boxplot(x='time', y='total_bill', data=df[df['time'] == 'Lunch'], ax=axes[1], color='green')
axes[1].set_title('Tổng Hoá Đơn - Lunch')
axes[1].set_xlabel('Thời Gian')

# Thêm tiêu đề cho toàn bộ figure
plt.suptitle('So Sánh Tổng Hoá Đơn Giữa Lunch và Dinner')

# Hiển thị đồ thị
plt.tight_layout()
plt.show()
````
