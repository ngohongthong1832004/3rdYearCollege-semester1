**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
import random
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.
def bai01_a():
    so_lan_kiem_tra = 10
    so_cau_hoi = 10
    so_cau_dung = 0
    for i in range(so_lan_kiem_tra):
        for j in range(so_cau_hoi):
            if random.randint(0, 3) == 0:
                so_cau_dung += 1
        print(f'Số câu đúng của bạn sinh viên trong lần kiểm tra thứ {i + 1} là: {so_cau_dung}')
        so_cau_dung = 0

# b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm
def bai01_b():
    so_lan_kiem_tra = 10
    so_cau_hoi = 10
    so_cau_dung = 0
    so_diem = 0
    for i in range(so_lan_kiem_tra):
        for j in range(so_cau_hoi):
            if random.randint(0, 3) == 0:
                so_cau_dung += 1
        so_diem += so_cau_dung * 4 - (so_cau_hoi - so_cau_dung) * 2
        so_cau_dung = 0
    print(f'Xác suất để sinh viên này được 4 điểm là: {so_diem / (so_lan_kiem_tra * 40)}')
    
# c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)
def bai01_c():
    so_lan_kiem_tra = 10
    so_cau_hoi = 10
    so_cau_dung = 0
    so_cau_dung_arr = []
    for i in range(so_lan_kiem_tra):
        for j in range(so_cau_hoi):
            if random.randint(0, 3) == 0:
                so_cau_dung += 1
        so_cau_dung_arr.append(so_cau_dung)
        so_cau_dung = 0
    print(f'E(X) = {np.mean(so_cau_dung_arr)}')
    print(f'Var(X) = {np.var(so_cau_dung_arr)}')
    
bai01_a()
bai01_b()
bai01_c()
```

**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
def bai02():
    lam = 3
    x = [1, 2, 3]
    p = []
    for i in x:
        p.append(np.exp(-lam) * lam ** i / np.math.factorial(i))
    print(f'Xác suất để trung tâm nhận được 1 cuộc gọi trong vòng một phút là: {p[0]}')
    print(f'Xác suất để trung tâm nhận được 2 cuộc gọi trong vòng một phút là: {p[1]}')
    print(f'Xác suất để trung tâm nhận được 3 cuộc gọi trong vòng một phút là: {p[2]}')

bai02()
```

**Bài 03:** Trọng lượng (đơn vị gam) của một loại trái cây có phân phối chuẩn với µ = 500 (gam) gam và $σ^2 = 16 (gam^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

Loại 1: trên 505 gam

Loại 2: từ 495 – 505 gam

Loại 3: dười 495 gam

a. Hãy tính tỷ lệ của mỗi loại.
b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

```python
def bai03():
    # a. Hãy tính tỷ lệ của mỗi loại.
    mu = 500
    sigma = 4
    p1 = 1 - (mu - 505) / sigma
    p2 = (mu + 5 - mu + 5) / sigma
    p3 = (mu - 5 - mu) / sigma
    print(f'Tỷ lệ của loại 1 là: {p1}')
    print(f'Tỷ lệ của loại 2 là: {p2}')
    print(f'Tỷ lệ của loại 3 là: {p3}')
    
    # b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
    so_lan_lay_mau = 20
    trai_cay = np.random.normal(mu, sigma, so_lan_lay_mau)
    print(f'Giá trị trung bình của trái cây từ mẫu là: {np.mean(trai_cay)}')
    print(f'Độ lệch chuẩn của trái cây từ mẫu là: {np.std(trai_cay)}')
    plt.hist(trai_cay, bins=10)
    plt.show()
    
    # c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
    so_lan_lay_mau = 100
    trai_cay = np.random.normal(mu, sigma, so_lan_lay_mau)
    print(f'Giá trị trung bình của trái cây từ mẫu là: {np.mean(trai_cay)}')
    print(f'Độ lệch chuẩn của trai cây từ mẫu là: {np.std(trai_cay)}')
    plt.hist(trai_cay, bins=10)
    plt.show()
    
bai03()

```

**Bài 6**

Giả sử X là biến ngẫu nhiên rời rạc

a. Hãy cho biết ý nghĩa câu lệnh sau: `binom.pmf(4, 10, 0.6)`

b. Giả sử X có phân phối nhị thức `X ~ B(10, 0.6)`. Hãy tính các giá trị sau:  
$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

c. Hãy cho biết ý nghĩa câu lệnh sau: `poisson.pmf(4, 3)`

d. Giả sử X có phân phối Poisson `X ~ P(3)`. Hãy tính các giá trị sau:  
$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

```python
def bai06():
    from scipy.stats import binom, poisson
    
    # a. Hãy cho biết ý nghĩa câu lệnh sau: `binom.pmf(4, 10, 0.6)`
    print(f'Ý nghĩa của câu lệnh binom.pmf(4, 10, 0.6) là: {binom.pmf(4, 10, 0.6)}')
    
    # b. Giả sử X có phân phối nhị thức `X ~ B(10, 0.6)`. Hãy tính các giá trị sau:  
    # $P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $
    print(f'P(X <= 5) = {binom.cdf(5, 10, 0.6)}')
    print(f'P(X < 5) = {binom.cdf(4, 10, 0.6)}')
    print(f'P(X >= 4) = {1 - binom.cdf(3, 10, 0.6)}')
    print(f'P(X = 5) = {binom.pmf(5, 10, 0.6)}')
    
    # c. Hãy cho biết ý nghĩa câu lệnh sau: `poisson.pmf(4, 3)`
    print(f'Ý nghĩa của câu lệnh poisson.pmf(4, 3) là: {poisson.pmf(4, 3)}')
    
    # d. Giả sử X có phân phối Poisson `X ~ P(3)`. Hãy tính các giá trị sau:  
    # $P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $
    print(f'P(X <= 5) = {poisson.cdf(5, 3)}')
    print(f'P(X < 5) = {poisson.cdf(4, 3)}')
    print(f'P(X >= 4) = {1 - poisson.cdf(3, 3)}')
    print(f'P(X = 5) = {poisson.pmf(5, 3)}')
    
bai06()
```

**Bài 7**

a. Cho biết ý nghĩa của các câu lệnh sau:

- `norm.cdf(2)`

- `norm.cdf(2, 1, 1)`

- `norm.cdf(2, 1, 2)`

b. Cho biết kết quả của câu lệnh sau: `norm.ppf(norm.cdf(2))`

c. Tính các giá trị sau và vẽ hình minh họa cho các giá trị tính được:

- `norm.ppf(0.975)`

- `norm.ppf(0.975, 1, 1)`

- `norm.ppf(0.975, 1, 2)`

d. Giả sử biến ngẫu nhiên X có phân phối chuẩn với kỳ vọng là 24 và phương sai là 16. Hãy tính các giá trị sau:

- $P(X \leq 20)$
- $P(X \ge 29.5)$
- $P(X = 23.8)$

```python
def bai07():
    from scipy.stats import norm
    
    # a. Cho biết ý nghĩa của các câu lệnh sau:
    # - `norm.cdf(2)`
    print(f'Ý nghĩa của câu lệnh norm.cdf(2) là: {norm.cdf(2)}')
    
    # - `norm.cdf(2, 1, 1)`
    print(f'Ý nghĩa của câu lệnh norm.cdf(2, 1, 1) là: {norm.cdf(2, 1, 1)}')
    
    # - `norm.cdf(2, 1, 2)`
    print(f'Ý nghĩa của câu lệnh norm.cdf(2, 1, 2) là: {norm.cdf(2, 1, 2)}')
    
    # b. Cho biết kết quả của câu lệnh sau: `norm.ppf(norm.cdf(2))`
    print(f'Kết quả của câu lệnh norm.ppf(norm.cdf(2)) là: {norm.ppf(norm.cdf(2))}')
    
    # c. Tính các giá trị sau và vẽ hình minh họa cho các giá trị tính được:
    # - `norm.ppf(0.975)`
    print(f'norm.ppf(0.975) = {norm.ppf(0.975)}')
    
    # - `norm.ppf(0.975, 1, 1)`
    print(f'norm.ppf(0.975, 1, 1) = {norm.ppf(0.975, 1, 1)}')
    
    # - `norm.ppf(0.975, 1, 2)`
    print(f'norm.ppf(0.975, 1, 2) = {norm.ppf(0.975, 1, 2)}')
    
    # d. Giả sử biến ngẫu nhiên X có phân phối chuẩn với kỳ vọng là 24 và phương sai là 16. Hãy tính các giá trị sau:
    # - $P(X \leq 20)$
    print(f'P(X <= 20) = {norm.cdf(20, 24, 16 ** 0.5)}')
    
    # - $P(X \ge 29.5)$
    print(f'P(X >= 29.5) = {1 - norm.cdf(29.5, 24, 16 ** 0.5)}')
    
    # - $P(X = 23.8)$
    print(f'P(X = 23.8) = {norm.pdf(23.8, 24, 16 ** 0.5)}')
    
bai07()
```

**Bài 8**

Một công ty sản xuất làm ra các chi tiết máy với độ dài các chi tiết có phân phối chuẩn với kỳ vọng là 3000 mm và độ lệch chuẩn là 3 mm. Yêu cầu cho các chi tiết máy này là phải có chiều dài từ 2993 mm đến 3007 mm. Tính tỷ lệ sản phẩm lỗi của công ty

```python
from scipy.stats import norm
def bai08():
    mu = 3000
    sigma = 3
    p = norm.cdf(3007, mu, sigma) - norm.cdf(2993, mu, sigma)
    print(f'Tỷ lệ sản phẩm lỗi của công ty là: {1 - p}')
    
bai08()
```

a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOIEC của một nhóm gồm 50 người.

b. Tính điểm trung bình và độ lệch chuẩn về điểm của nhóm.

c. Tìm miền giá trị, và miền phân vị (IQR) của nhóm

d. Cho biết tỷ lệ đạt trên 450 điểm của nhóm. 

e. Vẽ đồ thị histogram của nhóm và so sánh với đồ thị hàm mật độ của phân phối chuẩn với kỳ vọng là 500 và độ lệch chuẩn là 8.

f. Thay vì mô phỏng 50 người, hãy mô phỏng nhóm với 100 người thi. Bạn có nhận xét gì?

```python
def bai09():
    mu = 500
    sigma = 8
    so_luong_nguoi = 50
    diem_thi = np.random.normal(mu, sigma, so_luong_nguoi)
    print(f'Điểm trung bình của nhóm là: {np.mean(diem_thi)}')
    print(f'Độ lệch chuẩn của nhóm là: {np.std(diem_thi)}')
    print(f'Miền giá trị của nhóm là: {np.min(diem_thi)} - {np.max(diem_thi)}')
    print(f'Miền phân vị của nhóm là: {np.percentile(diem_thi, 75) - np.percentile(diem_thi, 25)}')
    print(f'Tỷ lệ đạt trên 450 điểm của nhóm là: {np.sum(diem_thi > 450) / so_luong_nguoi}')
    plt.hist(diem_thi, bins=10)
    plt.show()
    
    so_luong_nguoi = 100
    diem_thi = np.random.normal(mu, sigma, so_luong_nguoi)
    print(f'Điểm trung bình của nhóm là: {np.mean(diem_thi)}')
    print(f'Độ lệch chuẩn của nhóm là: {np.std(diem_thi)}')
    print(f'Miền giá trị của nhóm là: {np.min(diem_thi)} - {np.max(diem_thi)}')
    print(f'Miền phân vị của nhóm là: {np.percentile(diem_thi, 75) - np.percentile(diem_thi, 25)}')
    print(f'Tỷ lệ đạt trên 450 điểm của nhóm là: {np.sum(diem_thi > 450) / so_luong_nguoi}')
    plt.hist(diem_thi, bins=10)
    plt.show()
    
bai09()
```

**Bài 10**

Viết một chương trình để mô phỏng việc tung đồng xu và đánh giá phương sai của giá trị mô phỏng của biến ngẫu nhiên X (tung được mặt `Head`) đối với các số lần tung khác nhau. Giả sử  đồng xu phải cân bằng. 

Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó. 

Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

```python
def bai10():
    N = [10, 40, 90, 160, 250, 490, 640, 810, 1000]
    phuong_sai = []
    for n in N:
        x = []
        for i in range(100):
            x.append(np.sum(np.random.randint(0, 2, n)))
        phuong_sai.append(np.var(x))
    plt.plot(N, phuong_sai)
    plt.show()
    
bai10()
```
