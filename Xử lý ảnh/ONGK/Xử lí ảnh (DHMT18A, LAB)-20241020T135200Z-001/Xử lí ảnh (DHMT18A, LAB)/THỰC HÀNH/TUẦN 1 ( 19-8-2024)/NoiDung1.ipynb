```python
crop = image[100:200, 100:200]
plt.imshow(crop, cmap='gray')
plt.axis('off')
plt.show()
```

```python
temp = image.copy()
temp[100:200, 100:200] = 0
plt.imshow(temp, cmap='gray')
plt.axis('off')
plt.show()
```

```python
h, w = image.shape[:2]
ratio = h/w
resize_image = cv2.resize(image, (300, int(300*ratio)))
plt.imshow(resize_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
gray_16bit = image // 16
gray_8bit = image // 8
gray_4bit = image // 4
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(gray_16bit, cmap='gray')
plt.title('16 bit')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gray_8bit, cmap='gray')
plt.title('8 bit')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(gray_4bit, cmap='gray')
plt.title('4 bit')
plt.axis('off')
plt.show()
```
