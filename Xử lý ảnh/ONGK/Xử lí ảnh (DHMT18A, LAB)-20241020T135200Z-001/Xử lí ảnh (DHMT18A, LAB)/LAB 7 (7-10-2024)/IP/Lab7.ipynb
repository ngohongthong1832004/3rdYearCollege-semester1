```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def show(image):
  # plt.figure(figsize=(10,10))
  plt.imshow(image, cmap='gray')
  plt.axis('off')
  plt.show()
```

```python
image = cv2.imread('Image/brain.jpg', cv2.IMREAD_GRAYSCALE)
show(image)
```

# Otsu's Thresholding

```python
retval, otsu = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
show(otsu)
```

# Mean C-based thresholding & Gaussian Weighted Average thresholding




```python
dst1 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
cv2.THRESH_BINARY, 11, 2)
show(dst1)
```

```python
dst2 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
cv2.THRESH_BINARY, 11, 2)
show(dst2)
```

# Different values of type paramater

```python
dst1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)[1]
dst2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)[1]
dst3 = cv2.threshold(image, 127, 255, cv2.THRESH_TRUNC)[1]
dst4 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO)[1]
dst5 = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)[1]
show(dst1)
show(dst2)
show(dst3)
show(dst4)
show(dst5)
```
