```python
#import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading image
source = '/content/drive/MyDrive/Xử lí ảnh (DHMT18A, LAB)/Lab 1 (8-8-2024)/image.png'
image = cv2.imread(source)
image
```

```python
#Size of image
print(image.shape)
print(image.size)
print(image.dtype)
```

```python
#Showing image as BRG
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
#Display RGB image
# plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.imshow(image[:,:,::-1])
plt.axis('off')
plt.show()
```

```python
#Create a random image
array = np.random.randint(0, 256, 120000)
grayImage = array.reshape(300, 400)
plt.imshow(grayImage, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#Convert to RGB img
colorImage = grayImage.reshape(100, 400, 3)
plt.imshow(colorImage)
plt.axis('off')
plt.show()
```

```python
#Crop image
crop = image[100:500, 100:500]
plt.imshow(crop)
plt.axis('off')
plt.show()
```

```python
image = image[:,:,::-1]
```

```python
#Changing image color
copy = image.copy()
copy[100:500, 100:500] = [255, 255, 255]
plt.imshow(copy)
plt.axis('off')
plt.show()
```

```python
#Resizing images
resize = cv2.resize(image, (200, 200))
plt.imshow(resize)
plt.axis('off')
plt.show()
```

```python
#Resize while keeping aspect ratio
h, w = image.shape[:2]
ratio = h / w
dim = (1000, int(1000 * ratio))
resize = cv2.resize(image, dim)
plt.imshow(resize)
plt.axis('off')
plt.show()
```

```python
#Decrease gray level
grayImg = image[]
img_xam_4bit = (image // 16)  # Giảm xuống 16 mức xám (4 bit)
img_xam_2bit = (image // 64) # Giảm xuống 4 mức xám (2 bit)
img_xam_1bit = (image // 128) # Giảm xuống 2 mức xám (1 bit)

# Hiển thị các hình ảnh
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(1, 4, 2)
plt.imshow(img_xam_4bit, cmap='gray')
plt.title('Ảnh xám 4 bit')

plt.subplot(1, 4, 3)
plt.imshow(img_xam_2bit, cmap='gray')
plt.title('Ảnh xám 2 bit')

plt.subplot(1, 4, 4)
plt.imshow(img_xam_1bit, cmap='gray')
plt.title('Ảnh xám 1 bit')

plt.show()
```
