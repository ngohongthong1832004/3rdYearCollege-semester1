```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
trump = cv2.imread('trump.jpg')
trump = trump[:, :, ::-1]
moustache = cv2.imread('24883.png', -1)
```

```python
trump.shape
```

```python
moustache.shape
```

```python
moustacheRGB = moustache[:, :, 0:3]
moustacheAlpha = moustache[:, :, 3]
```

```python
moustacheMask = cv2.merge((moustacheAlpha, moustacheAlpha, moustacheAlpha))
moustacheMask = np.uint8(moustacheMask / 255)
```

```python
roi = trump[0:256, 0:256,:]
print (roi.dtype)
print(moustacheMask.dtype)
maskRoi = cv2.multiply(roi, (1 - moustacheMask))
```

```python
maskMoustache = cv2.multiply(moustacheRGB, moustacheMask)
```

```python
result = cv2.add(maskRoi, maskMoustache)
plt.imshow(result)
plt.axis('off')
plt.show()
```
