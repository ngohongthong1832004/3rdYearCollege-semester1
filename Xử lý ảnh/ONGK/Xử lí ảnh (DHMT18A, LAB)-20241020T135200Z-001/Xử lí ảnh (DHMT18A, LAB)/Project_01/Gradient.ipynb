```python
#Tạo một ma trận full chỉ số màu là 255
arr = np.full((256, 256), 255, dtype=np.uint8)
arr
```

```python
plt.imshow(arr, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#Lần lượt thay đổi các hàng có chỉ số từ 254 trở xuống
row = 1
for i in range(254, -1, -1):
  arr[row, :] = i
  row += 1
```

```python
plt.imshow(arr, cmap='gray')
plt.axis('off')
plt.show()
```
