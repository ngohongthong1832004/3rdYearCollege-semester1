```python
#Hàm vẽ đường thẳng từ pixel có tạo độ (x1, y1) tới pixel (x2, y2)
def drawLine(x1, y1, x2, y2):
  tan = (x2 - x1)/(y2 - y1)

  for i in range(x1, x2 + 1, -1):
    y = tan * (i) + x1
    image[i, int(y)] = 0
```

```python
drawLine(100, 0, 0, 100)
```

```python
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#Vẽ chồng 15 đường như thế sát nhau
for i in range (1, 16):
  drawLine(100 - i, 0, 0, 100 - i)
```

```python
image = cv2.imread('/content/drive/MyDrive/Xử lí ảnh (DHMT18A, LAB)/Project_01/image/img_3.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image
```

```python
def drawLine(x1, y1, x2, y2):
  tan = (x2 - x1)/(y2 - y1)
  for i in range(x1, x2 + 1, -1):
    y = tan * (i) + x1
    image[i, int(y), :] = 0


for i in range (0, 30):
  drawLine(200 - i, 0, 0, 200 - i)


plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```
