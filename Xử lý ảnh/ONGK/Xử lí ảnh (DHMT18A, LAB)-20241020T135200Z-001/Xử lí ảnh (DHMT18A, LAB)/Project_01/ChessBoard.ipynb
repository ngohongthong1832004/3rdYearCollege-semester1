```python

# black = np.array([[0]*4] * 4)
# white  = np.array([[1] * 4] * 4)

white = np.ones((4, 4), dtype=np.uint8)
# white = np.full((4, 4), 255)
black = np.zeros((4, 4), dtype=np.uint8)
white
```

```python
row1 = np.hstack((black, white) * 4)
row2 = np.hstack((white, black) * 4)
row1
```

```python
chessboard = np.vstack((row1, row2) * 4)
chessboard
```

```python
plt.imshow(chessboard, cmap='gray')
plt.axis('off')
plt.show()
```
