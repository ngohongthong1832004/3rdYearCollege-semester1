```python
h3 = np.array([[-1, -1, -1],
              [-1, 8, -1],
              [-1, -1, -1]]) #Edge-detection
h1 = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]]) #no-change
h2 = np.array([[0, 0, 0], [0, 0, 0], [0, 0, 1]]) #only change pixel (2, 2)
h4 = 1/16 * np.array([[1, 2, 1], [2, 4, 2], [1, 2, 1]]) #Gaussian filter
```

```python
result = cv2.filter2D(img, -1, h3)
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```

```python
result = cv2.filter2D(img, -1, h4)
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```

# EX 2.9 (Sharpening Image)

```python
img = cv2.imread('download.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
blur = cv2.medianBlur(img, 5)
```

```python
#step 1: Subtract img and blur img
edge = cv2.subtract(img, blur)
#step 2: Adding original img to edge
result = cv2.add(img, edge * 3)
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```
